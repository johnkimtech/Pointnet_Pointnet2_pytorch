```python
# make scale information for each point cloud
import os
from os.path import join, basename
import pandas as pd
from pathlib import Path
from glob import glob
from data_utils.helpers import read_point_cloud_text
from data_utils.ModelNetDataLoader import pc_normalize

ROOT_DIR = "data/modelnet40_withFootLast_LeftRight_AugFlip"
subdirs = ["foot_left", "last_left", "foot_right", "last_right"]
pc_files = []
for subdir in subdirs:
    search = Path(join(ROOT_DIR, subdir)) / "*txt"
    pc_files += glob(str(search))

df = pd.DataFrame(
    data={
        "file": map(basename, pc_files),
        "scale": map(
            lambda pc_file: pc_normalize(
                read_point_cloud_text(pc_file), return_scale=True
            )[1],
            pc_files,
        ),
    }
)
```

```python
scale_csv = df.to_csv(Path(ROOT_DIR) / "scale.csv", index=False)
```
